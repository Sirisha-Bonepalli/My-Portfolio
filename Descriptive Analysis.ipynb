```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Enable inline plotting for Jupyter Notebook
%matplotlib inline

# Load the dataset
dataset_path = r"C:\Users\ASUS\Downloads\Sirisha Property tax Dataset.csv"
df = pd.read_csv(dataset_path)

# Display basic information
display("Dataset Overview:")
display(df.info())
```

```python
display("\nSummary Statistics:")
display(df.describe())

display("\nMissing Values:")
display(df.isnull().sum())
```

```python
# Checking unique values for categorical variables
categorical_features = df.select_dtypes(include=['object']).columns
for feature in categorical_features:
    display(f"Unique values in {feature}:")
    display(df[feature].unique())
```

```python
# Checking numerical statistics
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
if numeric_cols:
    display("\nDetailed Statistics for Numerical Features:")
    display(df[numeric_cols].describe(include='all'))
else:
    display("No numerical features found.")
```

```python
# Distribution plots for numerical features
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# Boxplot for outlier detection
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[numeric_cols])
plt.xticks(rotation=90)
plt.title("Boxplot of Numerical Features")
plt.show()
```
